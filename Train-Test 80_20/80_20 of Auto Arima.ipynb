The null hypothesis (H₀) of the Ljung-Box test states that the residuals are white noise (i.e., no significant autocorrelation).
The alternative hypothesis (H₁) suggests that the residuals are autocorrelated (indicating a poor model fit).
A p-value > 0.05 means we fail to reject the null hypothesis, implying that the residuals behave like white noise.
This suggests that the ARIMA(1,1,0) model captures most of the autocorrelated structure in the data well.

## Checking The Assumption

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Split the data into training and test sets
train_size = int(len(data['Inflation']) * 0.8)
train, test = data['Inflation'][:train_size], data['Inflation'][train_size:]

# Fit ARIMA(4,1,2)(0,0,1)[12] model
model_412_001_12 = ARIMA(train, order=(4, 1, 2), seasonal_order=(0, 0, 1, 12))
model_412_001_12_fit = model_412_001_12.fit()

# Residual Analysis
residuals = model_412_001_12_fit.resid

# 1. Check for White Noise
plt.figure(figsize=(12,5))
plt.plot(residuals)
plt.axhline(y=0, linestyle='dashed', color='black')
plt.title('Residuals Over Time (White Noise Check)')
plt.show()

# ACF and PACF plots
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# ACF plot
sm.graphics.tsa.plot_acf(residuals, lags=40, ax=ax[0])
ax[0].set_title('Autocorrelation Function (ACF)')

# PACF plot
sm.graphics.tsa.plot_pacf(residuals, lags=40, ax=ax[1])
ax[1].set_title('Partial Autocorrelation Function (PACF)')

plt.tight_layout()  # Ensures the plots do not overlap
plt.show()

# Ljung-Box Test
ljung_box_results = acorr_ljungbox(residuals, lags=[10], return_df=True)
print("Ljung-Box test p-value:", ljung_box_results['lb_pvalue'].values[0])

# 3. Normality of Residuals (Histogram and Q-Q Plot)
fig, ax = plt.subplots(1, 2, figsize=(14,5))
sns.histplot(residuals, bins=30, kde=True, ax=ax[0])
ax[0].set_title("Histogram of Residuals")
sm.qqplot(residuals, line='s', ax=ax[1])
ax[1].set_title("Q-Q Plot of Residuals")
plt.show()

from scipy.stats import shapiro

# Perform Shapiro-Wilk test
shapiro_test_stat, shapiro_p_value = shapiro(residuals)

# Print results
print("Shapiro-Wilk test statistic:", shapiro_test_stat)
print("Shapiro-Wilk test p-value:", shapiro_p_value)

# Interpretation
if shapiro_p_value > 0.05:
    print("Residuals appear to be normally distributed (fail to reject H0).")
else:
    print("Residuals do not appear to be normally distributed (reject H0).")

# Plot residuals vs. fitted values
plt.figure(figsize=(10, 6))
plt.scatter(model_412_001_12_fit.fittedvalues, residuals, alpha=0.5)
plt.axhline(y=0, color="red", linestyle="--")
plt.title("Residuals vs. Fitted Values")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

# Compute evaluation metrics
pred_412_001_12 = model_412_001_12_fit.predict(start=len(train), end=len(train)+len(test)-1)
mse_412_001_12 = mean_squared_error(test, pred_412_001_12)
rmse_412_001_12 = np.sqrt(mse_412_001_12)
mae_412_001_12 = mean_absolute_error(test, pred_412_001_12)
mape_412_001_12 = np.mean(np.abs((test - pred_412_001_12) / test)) * 100
aic_412_001_12 = model_412_001_12_fit.aic
bic_412_001_12 = model_412_001_12_fit.bic

# Print evaluation metrics
print("ARIMA(4,1,2)(0,0,1)[12] Metrics:")
print(f"MSE: {mse_412_001_12}, RMSE: {rmse_412_001_12}, MAE: {mae_412_001_12}, MAPE: {mape_412_001_12}, AIC: {aic_412_001_12}, BIC: {bic_412_001_12}")

```
